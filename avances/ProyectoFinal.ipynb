```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="deep")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
data = pd.read_csv('Dataset.csv')

data.info()
```

```python
data['SERVICIO'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Servicios")
plt.xlabel("Servicio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(20,8))

sns.histplot(data['CLASIFICACION'], kde=True)

plt.xlabel("Clasificación")
plt.ylabel("Frecuencia")
plt.title("Distribución de Clasificación")

plt.xticks(rotation=35)

plt.show()
```

```python
tabla = pd.crosstab(data['DEPARTAMENTO_PRESTACION'], data['SERVICIO'])

sns.heatmap(tabla, cmap="Blues")
plt.title("Relación Servicio vs Departamento de Prestación")
plt.ylabel("Departamento")
plt.xlabel("Servicio")
plt.show()
```

```python
top_dept = data['DEPARTAMENTO_PRESTACION'].value_counts().index[:5]

filtro = data[data['DEPARTAMENTO_PRESTACION'].isin(top_dept)]

sns.countplot(data=filtro,
              x='DEPARTAMENTO_PRESTACION',
              hue='SERVICIO')
plt.title("Servicios por Departamento (Top 5)")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=data, x='SERVICIO', y='CLASIFICACION')
plt.xticks(rotation=45)
plt.title("Variación de Clasificación según Servicio")
plt.show()

```
